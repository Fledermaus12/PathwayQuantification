```python
import pandas as pd 
import numpy as np
flock = pd.read_excel('FlockhardtTable.xlsx')
flock
```

```python
drugs = list()
var1 = list(flock['2C9'].dropna())
drugs.extend(var1)
var2 = list(flock['2C19'].dropna())
drugs.extend(var2)
var3 = list(flock['2D6'].dropna())
drugs.extend(var3)
print(drugs)
drugs = list(set(drugs))

li2c9 = list()
li2c19 = list()
li2d6 = list()
for drug in drugs:
    if drug in var1: li2c9.append(1)
    else: li2c9.append(0)

    if drug in var2: li2c19.append(1)
    else: li2c19.append(0)

    if drug in var3: li2d6.append(1)
    else: li2d6.append(0)

flock_conv = pd.DataFrame({
    'drugs': drugs,
    '2d6sub': li2d6,
    '2c19sub': li2c19,
    '2c9sub': li2c9
})
flock_conv
```

```python
flock_conv.to_excel('FlockhardtTable_converted.xlsx')
```
