Prepare the functions to convert the modified FDA table. The FDA Table has been reduced to one column. Numbers in medications names are manually deleted. 

```python

fda_table = pd.read_excel('fda_table_clean.xlsx')
fda_table = fda_table.set_index('Drug or Other Substance')
fda_table
```

```python
s = fda_table.loc['voriconazole', 'CYP SENS SUB']
if '2C9' in s:
    print('true')
else: print('false')
```

```python
output_list = []
for drug in fda_table.index:
    enzyme_list = []
    s = fda_table.loc[drug, 'CYP SENS SUB']
    if '2C9' in s:
        enzyme_list.append('2c9sub')
    if '2C19' in s:
        enzyme_list.append('2c19sub')
    if '2D6' in s:
        enzyme_list.append('2d6sub')
    output_list.append(enzyme_list)
fda_table['interactions'] = output_list
fda_table
```

```python
def search(keyword, liste):
    '''
    :param keyword: the enzyme name, that has to be sorted out
    :param liste: the list of the identified interactions
    :return: a list with 1 & 0, which entry has shown a interaction with given enzyme (keyword)
    '''
    lkeyword = list()
    for elem in liste:
        if keyword in elem: lkeyword.append(1)
        else: lkeyword.append(0)
    return lkeyword
enzymes = ['2d6sub', '2c19sub', '2c9sub']
for i in enzymes:
    search_result = search(i, list(fda_table['interactions']))
    fda_table[i] = search_result
fda_table
```

```python
fda_table.to_excel('fda_table_clean_converted.xlsx')
```
